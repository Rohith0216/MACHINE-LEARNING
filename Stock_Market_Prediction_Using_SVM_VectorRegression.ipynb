Importing Libraries

```python
import pandas as pd
import numpy as np
```

 Load Dataset

```python
dataset=pd.read_csv('stock--market.csv')
```

Summarise Dataset

```python
print(dataset.shape)
print(dataset.head(5))
dataset
```

Segregate Value Into input X & Output y

```python
x=dataset.iloc[:,:-1].values
x
```

```python
y=dataset.iloc[:,-1].values
y
```

Spliting Dataset For Testing Our Model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

Training Dataset Using VectorRegression

```python
from sklearn.svm import SVR
model =SVR()
model.fit(x_train,y_train)

```

Predicition For All Test Data For Validation

```python
y_pred=model.predict(x_test)
from sklearn.metrics import r2_score ,mean_squared_error
mse= mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print("mse error",rmse)
r2score=r2_score(y_test,y_pred)
print('r2score',r2score*100)
```
